```python
import pandas as pd
```

```python
df= pd.read_csv('heart.csv')
```

```python
#Check if dataset is loaded and analyze for missing values or outliers
df.head
```

```python
#data set looks clean with no missing values
df.isnull()
```

```python
#splitting data into 70-30 to train and test the model
from sklearn.model_selection import train_test_split

X=df[['age', 'sex', 'cp', 'trestbps', 'chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']]  # Features 
y=df['target']  # Labels

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
#import random forest classifier using sklearn
from sklearn.ensemble import RandomForestClassifier

#Create a Random forest Classifier all default values and try various number of tree such that we get a better accuracy and model is not underfittingor overfitting
clf=RandomForestClassifier(n_estimators=50)
#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
```

```python
#check for the accuracy of the model
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#splitting data into 70-30 to train and test the model

from sklearn.model_selection import train_test_split

X=df[['age', 'sex', 'cp', 'trestbps', 'chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']]  # Features 
y=df['target']  # Labels

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
gnb = GaussianNB()

#Train the model using the training sets
gnb.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = gnb.predict(X_test)
```

```python
#check for the accuracy of the model
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
